```python
url = 'http://cocl.us/Geospatial_data'
```

```python
LLData = pd.read_csv(url, header=0)
```

```python
LLData.head()
```

```python
TEMP = pd.DataFrame(columns=['Postal Code', 'Latitude', 'Longitude'])
TEMP
```

```python
for index, row in CanData.iterrows():
    a = (LLData.loc[LLData['Postal Code'] == CanData['Postcode'][index]])
    TEMP = TEMP.append(a, ignore_index=True)
    CanData['Latitude'][index]=TEMP['Latitude'][index]
    CanData['Longitude'][index]=TEMP['Longitude'][index]
```
