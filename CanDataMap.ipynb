```python
TEMP = pd.DataFrame(columns=['Postcode', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude'])
TEMP
```

```python
a = CanData.Borough.str.contains("Toronto")
```

```python
for index, row in CanData.iterrows():
    b = CanData.iloc[index]
    if a[index] == True:
        TEMP = TEMP.append(b, ignore_index=True)
```

```python
TEMP
```

```python
from sklearn.cluster import KMeans

kclusters = 5

manhattan_grouped_clustering = TEMP['Borough']

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(manhattan_grouped_clustering)

kmeans.labels_[0:10] 
```
